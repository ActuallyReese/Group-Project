```python
import pandas as pd
import os
```

```python
energy_data = "WorldEnergyConsumption.csv"
```

```python
energy_data_df = pd.read_csv(energy_data)
energy_data_df.head(10)
```

```python
energy_data_df.isnull()
```

```python
energy_data_df = energy_data_df.fillna(0)
```

```python
energy_data_df.head()
```

```python
energy_data_df.isnull()
```

```python
electricity_production = energy_data_df[["country", "year","electricity_generation", "gdp"]]
electricity_production.head()
```

```python
electricity_prod_avg = electricity_production.groupby(['country']).mean()
electricity_prod_avg.head(10)
```

```python
electricity_production.to_csv('Electricity_Generation_by_Country.csv', index=False, encoding='utf-8')
```

```python
electricity_prod_avg.to_csv('Average_Electricity_Generation_by_Country', index=False, encoding='utf-8')
```

```python
energy_prod_country = energy_data_df[["country", "year","coal_production", "coal_prod_per_capita","energy_per_gdp", "gdp", "energy_per_capita", "gas_production", "gas_prod_per_capita", "oil_production", "oil_prod_per_capita",]]
energy_prod_country
```

```python
energy_prod_country.to_csv('Energy_Production_by_Country', index=False, encoding='utf-8')
```
