```python
import pandas as pd
import os 
```

```python
energy_data_updated = "Updated_Energy_Consumption.csv"
energy_data_df = pd.read_csv(energy_data_updated)
energy_data_df.head(10)
```

```python
year = energy_data_df['year']
country = energy_data_df['country']
population = energy_data_df['population']
gdp = energy_data_df['gdp']
gdp.head(10)
```

```python
world_table = pd.DataFrame({
    "Country": country,
    "Year": year,
    "Population": population,
    "GDP": gdp,
})

world_table.head(10)
```

```python
world_table.to_csv('World_Table', index=False)
```

```python
biofuel_electricity = energy_data_df['biofuel_electricity']
coal_electricity = energy_data_df['coal_electricity']
gas_electricity = energy_data_df['gas_electricity']
hydro_electricity = energy_data_df['hydro_electricity']
nuclear_electricity = energy_data_df['nuclear_electricity']
oil_electricity = energy_data_df['oil_electricity']
solar_electricity = energy_data_df['solar_electricity']
wind_electricity = energy_data_df['wind_electricity']
wind_electricity.head(10)
```

```python
production_table = pd.DataFrame({
    "Country": country,
    "Year": year,
    "biofuel_electricity": biofuel_electricity,
    "gas_electricity": gas_electricity,
    "hydro_electricity": hydro_electricity,
    "nuclear_electricity": nuclear_electricity,
    "oil_electricity": oil_electricity,
    "solar_electricity": solar_electricity,
    "wind_electricity": wind_electricity,
})
```

```python
production_table.head(10)
```

```python
production_table.to_csv('Production_Table', index=False)
```

```python
biofuel_consumption = energy_data_df['biofuel_consumption']
coal_consumption = energy_data_df['coal_consumption']
gas_consumption = energy_data_df['gas_consumption']
hydro_consumption = energy_data_df['hydro_consumption']
nuclear_consumption = energy_data_df['nuclear_consumption']
oil_consumption = energy_data_df['oil_consumption']
solar_consumption = energy_data_df['solar_consumption']
wind_consumption = energy_data_df['wind_consumption']
```

```python
consumption_table = pd.DataFrame({
    "Country": country,
    "Year": year,
    "biofuel_consumption": biofuel_consumption,
    "coal_consumption": coal_consumption,
    "gas_consumption": gas_consumption,
    "hydro_consumption": hydro_consumption,
    "nuclear_consumption": nuclear_consumption,
    "oil_consumption": oil_consumption,
    "solar_consumption": solar_consumption,
    "wind_consumption": wind_consumption,
})
consumption_table.head(10)
```

```python
consumption_table.to_csv('Consumption_Table', index=False)
```
