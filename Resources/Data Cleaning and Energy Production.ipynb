```python
import pandas as pd
import os
```

```python
energy_data = "WorldEnergyConsumption.csv"
```

```python
energy_data_df = pd.read_csv(energy_data)
energy_data_df.head(10)
```

```python
# Checking for null values
energy_data_df.isnull()
```

```python
energy_data_clean = energy_data_df.fillna(0)
```

```python
energy_data_clean.head()
```

```python
# Dropping the iso_code column because it is redundant
energy_data_cleaner = energy_data_clean.drop(columns= 'iso_code')
```

```python
# Creating a DataFrame that only focuses on United States
energy_data_USA = energy_data_cleaner[energy_data_cleaner["country"] == "United States"]
energy_data_USA.head()
```

```python
fuel_production_USA = energy_data_USA[["country", "year", "coal_production", "gas_production", "oil_production", "gdp"]].drop(columns= 'country')
fuel_production_USA.to_csv('Fuel_Production_USA.csv', index=False)
```

```python
electricity_production = energy_data_clean[["country", "year","electricity_generation", "gdp"]]
electricity_production.head()
```

```python
electricity_prod_avg = electricity_production.groupby(['country']).mean()
electricity_prod_avg.head(10)
```

```python
electricity_production.to_csv('Electricity_Generation_by_Country.csv', index=False, encoding='utf-8')
```

```python
electricity_prod_avg.to_csv('Average_Electricity_Generation_by_Country.csv', index=False, encoding='utf-8')
```

```python
energy_prod_country = energy_data_clean[["country", "year","coal_production", "coal_prod_per_capita","energy_per_gdp", "gdp", "energy_per_capita", "gas_production", "gas_prod_per_capita", "oil_production", "oil_prod_per_capita",]]
energy_prod_country
```

```python
energy_prod_country.to_csv('Energy_Production_by_Country.csv', index=False, encoding='utf-8')
```
